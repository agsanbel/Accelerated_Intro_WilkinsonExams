```python
#insert the data in the table of database
import pymysql.cursors
import csv
connection = pymysql.connect(host='localhost',
                            user='root',
                            password='root',
                            db='exam2',
                            charset='utf8mb4', 
                            cursorclass=pymysql.cursors.DictCursor)
connection.autocommit = False

 

try:
    with connection.cursor() as cursor:
        with open('Germplasm.tsv') as csvfile:
            germplasm = csv.DictReader(csvfile, delimiter = '\t')
            for row in germplasm:
                print(row)
                sql = """INSERT INTO germplasm (Locus, germplasm, phenotype, pubmed) values ('{}', '{}', '{}', '{}')""".format(row["Locus"], row["germplasm"], row["phenotype"], row["pubmed"])
                cursor.execute(sql)
        connection.commit() 
finally: 
    connection.close()
```

```python
# selection of join of two tables
exe1 = %sql SELECT * FROM germplasm RIGHT JOIN locus ON germplasm.Locus = locus.Locus;
```

```python
#SElection of gene skor and maa3
exe2 = %sql SELECT * FROM germplasm RIGHT JOIN locus ON germplasm.Locus = locus.Locus WHERE locus.Gene = "SKOR" OR locus.Gene = "MAA3";
```

```python
#Selection by chromosome
exe3 = {}
for n in range(1,6):
    x = "AT" + str(n) + "%"
    exe = %sql SELECT COUNT(*) FROM germplasm RIGHT JOIN locus ON germplasm.Locus = locus.Locus WHERE germplasm.Locus LIKE :x ;  
    exe3[n] = exe
    print("Chromosome", n, exe)
print(exe3)
```

```python
#Average of the protein lenght and creating a dicc to solve the data
exe4 = {}
for n in range(1,6):
    x = "AT" + str(n) + "%"
    exe = %sql SELECT AVG(locus.ProteinLength) FROM germplasm RIGHT JOIN locus ON germplasm.Locus = locus.Locus WHERE germplasm.Locus LIKE :x ;  
    exe4[n] = exe
    print("Chromosome", n, exe)
print(exe4)
```

```python
result_exam2 = exe1 + exe2
exercise4_exam2 = open("exe4.csv", "w")
exercise4_exam2.write(str(result_exam2))
exercise4_exam2 = open("exe4.csv", "a")
exercise4_exam2.write(exe3)
exercise4_exam2.write(exe4)
exercise4_exam2.close()
```
